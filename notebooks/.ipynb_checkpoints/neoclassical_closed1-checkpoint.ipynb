# The Neoclassical Growth Model in Discrete Time: Closed Form - Example 1

## Introduction
There are two well known closed form solutions for the neoclassical growth model in discrete time. One of the most well-known examples might be the following neoclassical deterministic growth model as pedagogically discussed, for instance, in Ljungqvist and Sargent (2018):
\begin{align}
& \max_{\left\{c_{t}\right\}_{t=0}^{\infty}} \sum_{t=0}^{\infty}\beta^{t}\ln c_{t} \\
\text{s.t.} & k_{t+1} + c_{t} \leq Ak_{t}^{\alpha}
\end{align}
where we assume full capital depreciation. The above neoclassical growth model is known to have the following analytical solution (policy function):
\begin{align}
k_{t+1} = \beta \alpha A k_{t}^{\alpha}
\end{align}

## Time iteration collocation with the Gaussian process
In this notebook, we aim to approximate the policy function by using the time iteration collocation method, where the unknown policy functions are approximated by the Gaussian process.

```python
# Import necessary packages
import sys
import warnings
import numpy as np
import pyipopt
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import GPy
from matplotlib import rc
from mpl_toolkits.mplot3d import Axes3D
# Use TeX font
rc('font', **{'family': 'sans-serif', 'serif': ['Helvetica']})
rc('text', usetex=True)
plt.rcParams["font.size"] = 15

np.random.seed(0)  # Fix the seed to get the same result

%matplotlib inline
```

```python
# Parameter setting, no TFP shock
A = 1  # Technology level
alpha = 0.36  # Capital share in the Cobb-Douglas production function
beta = 0.99  # Discount factor
```

```python
# Analytical solution
def k_compute_infty(alpha, beta, A):
    """ Return the stationary point (or steady state) """
    return (1 / (beta * alpha * A))**(1/(alpha - 1))


k_infty = k_compute_infty(alpha, beta, A)
print("Stationary point is {:5f}".format(k_infty))


def kplus_compute_analytic(k, alpha=alpha, beta=beta, A=A):
    """ Analytical solution
    Return the optimal capital stock in the next period """
    return alpha * beta * A * k**alpha


def c_compute(k, kplus, alpha=alpha, beta=beta, A=A):
    """ Return the optimal consumption policy """
    return A*k**alpha - kplus
```

```python
# Set the capital domain
kbeg, kend = 1e-2, 0.25
gridsize = 250
kgrid = np.linspace(kbeg, kend, gridsize)

plt.plot(kgrid, kplus_compute_analytic(kgrid))
plt.plot(kgrid, kgrid, 'k:')
plt.xlabel(r"$k_{t}$")
plt.ylabel(r"$k_{t+1}$")
plt.xlim([kbeg, kend])
plt.show()

plt.plot(kgrid, c_compute(kgrid, kplus_compute_analytic(kgrid)))
plt.xlabel(r"$k_{t}$")
plt.ylabel(r"$c_{t}$")
plt.xlim([kbeg, kend])
plt.show()
```

When we solve the model for the first-order condition, we derive the following equilibrium conditions:
\begin{align}
& \lambda_{t} - \beta \lambda_{t+1}A\alpha k_{t+1}^{\alpha-1} = 0 \\
& Ak_{t}^{\alpha} - k_{t+1} - \frac{1}{\lambda_{t}} = 0
\end{align}
We solve the two equilibrium condition by the time iteration.

```python
# Set the number of training and test examples
num_train = 5
num_test = 1000
num_input = 1

# Time iteration collocation setting
num_iter = 50
epsilon_tol = 1e-6
```

```python
# Initialize the Gaussian process
# Training datasets, assume that the steady state values
x_train_k = np.random.uniform(kbeg, kend, (num_train, num_input))
y_train_kplus = x_train_k
y_train_lambd = 1 / (A*x_train_k**alpha - x_train_k)

# Test datasets
x_test_k = np.random.uniform(kbeg, kend, (num_test, num_input))
```

```python
# Set the kernel objects
kernel_kplus = GPy.kern.RBF(input_dim=num_input)
kernel_lambd = GPy.kern.RBF(input_dim=num_input)

gp_kplus = GPy.models.GPRegression(x_train_k, y_train_kplus, kernel_kplus)
gp_kplus.plot()
gp_kplus.optimize_restarts(verbose=False)
gp_kplus.plot()
# sys.exit(0)
gp_lambd = GPy.models.GPRegression(x_train_k, y_train_lambd, kernel_lambd)
gp_lambd.plot()
gp_lambd.optimize_restarts(verbose=False)
gp_lambd.plot()
```

As seen above, the Gaussian process with the optimized hyper-parameters nicely predict the posterior distribution of the value of the target outcomes.

Nest we approximate the above two equilibrium conditions by the time iteration collocation with the Gaussian process, where I use IPOPT to solve the system of non-linear equations.

```python
def euler(x0, k):
    """
    Solve the system of non-linear equilibrium conditions by IPOPT.
    Policy functions are approximated by the Gaussian process.
    x0: Starting point
    k: Current capital state
    x[0]: Capital stock in the next period
    x[1]: lambda, Lagrange multiplier associated with the budget constraint
    """
    nvar = 2  # Number of variables
    # Variables are assumed to be non-negative
    x_L = np.zeros(nvar, dtype=np.float64)
    x_U = np.ones(nvar, dtype=np.float64) * np.inf

    def lambdplus(x):
        """ Return lambdaplus (the Lagrange multiplier) in the next period """
        return gp_lambd.predict(x[0].reshape(-1, num_input))[0].ravel()

    def eval_f(x):
        """ Dummy objective function """
        return 0

    def eval_grad_f(x):
        "Gradient of the dummy objective function"
        grad_f = np.zeros(nvar, dtype=np.float64)
        return grad_f

    ncon = nvar  # Number of constraints

    """ Complementarity constraints
    g0: Euler equation wrt. k
    g1: Resource constraint """

    g_L = np.zeros(ncon, dtype=np.float64)
    g_U = g_L

    nnzj = int(nvar * ncon)  # Number of (possibly) non-zeros in Jacobian
    nnzh = int((nvar**2 - nvar) / 2 + nvar)  # Number of non-zeros in Hessian

    def eval_g(x):
        """ The system of non-linear equilibrium conditions """
        g0 = x[1] - beta * lambdplus(x) * A * alpha * x[0]**(alpha-1)
        g1 = A*k**alpha - x[0] - 1 / x[1]
        return np.array([g0, g1], dtype=np.float64)

    def eval_jac_g(x, flag):
        """ Numerical approximation of the Jacobian of the system of
        non-linear equilibrium conditions
        Use the finite-difference-values option provided by IPOPT """
        row_idx = np.empty(nnzj, dtype=int)  # Row index
        col_idx = np.empty(nnzj, dtype=int)  # Column index

        # Jacobian matrix structure
        if flag:
            for i in range(ncon):
                for j in range(nvar):
                    row_idx[j + i * nvar] = i
                    col_idx[j + i * nvar] = j

            return (row_idx, col_idx)

    # ----------------------------------------------------------------------- #
    # Define a NLP model
    # ----------------------------------------------------------------------- #
    pyipopt.set_loglevel(0)  # Let IPOPT quite

    bm1972 = pyipopt.create(
        nvar, x_L, x_U, ncon, g_L, g_U, nnzj, nnzh, eval_f, eval_grad_f,
        eval_g, eval_jac_g)
    bm1972.str_option("linear_solver", "ma57")
    bm1972.str_option("jacobian_approximation", "finite-difference-values")
    bm1972.str_option("hessian_approximation", "limited-memory")
    bm1972.int_option("print_level", 0)
    
    xstar, zl, zu, constraint_multipliers, obj, status = bm1972.solve(x0)

    if status != 0:
        warnings.warn(
            "IPOPT fails to solve the system of non-linear equations")
    return xstar
```

```python
# --------------------------------------------------------------------------- #
# Time iteration collocation
# --------------------------------------------------------------------------- #
epsilons = []  # Track the approximation error
for n in range(num_iter):
    # Starting value in the optimization
    # Retlieve from the previous iteration
    x0 = np.stack([y_train_kplus.ravel(), y_train_lambd.ravel()], axis=1)

    # Keep the optimal policies
    y_train_kplus = np.empty((num_train, num_input))
    y_train_lambd = np.empty_like(y_train_kplus)

    for idx, k in enumerate(x_train_k):
        # For each state, solve the system of non-linear equations
        xstar = euler(x0[idx, :], k)
        # Track the optimal policies
        y_train_kplus[idx, 0] = xstar[0]
        y_train_lambd[idx, 0] = xstar[1]

    # Train the optimal policy with the Gaussian process
    kernel_kplus_updated = GPy.kern.RBF(input_dim=num_input)
    gp_kplus_updated = GPy.models.GPRegression(
        x_train_k, y_train_kplus, kernel_kplus_updated)
    gp_kplus_updated.optimize_restarts(num_restarts=10, verbose=False)

    kernel_lambd_updated = GPy.kern.RBF(input_dim=num_input)
    gp_lambd_updated = GPy.models.GPRegression(
        x_train_k, y_train_lambd, kernel_lambd_updated)
    gp_lambd_updated.optimize_restarts(num_restarts=10, verbose=False)

    # ----------------------------------------------------------------------- #
    # Approximation error analysis and update the policy functions
    # ----------------------------------------------------------------------- #
    epsilon = max(
        np.max(np.abs(gp_kplus.predict(x_test_k)[0]
                      - gp_kplus_updated.predict(x_test_k)[0])),
        np.max(np.abs(gp_lambd.predict(x_test_k)[0]
                      - gp_lambd_updated.predict(x_test_k)[0]))
    )
    epsilons.append(epsilon)  # Track the history of epsilon
    if n % 1 == 0:
        print('Iteration: {}, current approximation error{:5e}'.format(
            n, epsilons[-1]))
    if epsilons[-1] < epsilon_tol:
        gp_kplus_star = gp_kplus_updated
        gp_lambd_star = gp_lambd_updated
        print("Time iteration collocation is terminated successfuly.")
        break  # Terminate the iteration
    else:
        # Update the policy functions for the next iteration
        gp_kplus = gp_kplus_updated
        gp_lambd = gp_lambd_updated
```

```python
# Plot the approximation error
plt.plot(epsilons)
plt.yscale('log')
plt.xlabel("Number of iterations")
plt.ylabel("Max approximation error")
plt.xlim([0, None])
plt.show()
```

```python
kplus_star = gp_kplus_star.predict(kgrid.reshape(gridsize, num_input))
mu_kplus_star = kplus_star[0]
sigma_kplus_star = np.sqrt(kplus_star[1])

mu_cstar = c_compute(kgrid, mu_kplus_star.ravel(), alpha=alpha, beta=beta, A=A)
```

```python
plt.plot(kgrid, mu_kplus_star, label=r"$\mu$")
plt.plot(kgrid, mu_kplus_star + 2*sigma_kplus_star , 'r:', label=r"$\mu+2\sigma$")
plt.plot(kgrid, mu_kplus_star - 2*sigma_kplus_star , 'r-.', label=r"$\mu-2\sigma$")
plt.plot(kgrid, kplus_compute_analytic(kgrid), label=r"True")
plt.plot(kgrid, kgrid, 'k:')
plt.xlabel(r"$k_{t}$")
plt.ylabel(r"$k_{t+1}$")
plt.xlim([kbeg, kend])
plt.legend(loc='best')
plt.show()

plt.plot(kgrid, mu_cstar, label=r"$\mu$")
plt.plot(kgrid, c_compute(kgrid, kplus_compute_analytic(kgrid)), label=r"True")
plt.xlabel(r"$k_{t}$")
plt.ylabel(r"$c_{t}$")
plt.xlim([kbeg, kend])
plt.legend(loc='best')
plt.show()
```
