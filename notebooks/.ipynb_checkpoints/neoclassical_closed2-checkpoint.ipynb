# Neoclassical growth model in discrete time: Closed Form - Example 2

## Introduction

There are two well known closed form solutions for the neoclassical growth model in discrete time. The first and might be the most well-known example is discussed, for instance, in Ljungqvist and Sargent (2018). In this notebook, we aim to solve the second example where the social planner maximizes his infinite time horizon utility subject to the budget constraint with full capital depreciation:
\begin{align}
& \max \sum_{t=0}^{\infty}\beta^{t}\ln \left(C_{t} -\psi N_{t}^{\theta}\right) \\
\text{s.t.} \quad & K_{t+1} + C_{t} \leq A K_{t}^{\alpha} N_{t}^{1-\alpha}
\end{align}
where we notice that the utility function follows the classical Greenwood–Hercowitz–Huffman preferences (Greenwood et al. (1988)).

The above neoclassical growth model is well known to have a closed form solution, composing of:
\begin{align}
& K_{t+1} = \alpha \beta A K_{t}^{\alpha} N_{t}^{1-\alpha} \\
& C_{t} = \left(1-\alpha \beta \right) A K_{t}^{\alpha} N_{t}^{1-\alpha}
\end{align}

The aim of this notebook is to numerically approximate the above policy functions based on the Gaussian process regression.

## Analytical solutions

As discussed above, we can compute the stationary point as well as plot the optimal policy functions. However before doing so, we import some necessary numerical packages and define the economic parameters. Note that we use the GPy to regress the optimal policy functions based on the Gaussian process:

```python
import sys
import warnings
import numpy as np
import pyipopt
import GPy
from scipy.optimize import root
# plot
import matplotlib.pyplot as plt
from matplotlib import rc
from mpl_toolkits.mplot3d import Axes3D
# Use TeX font
rc('font', **{'family': 'sans-serif', 'serif': ['Helvetica']})
rc('text', usetex=True)
plt.rcParams["font.size"] = 15

np.random.seed(0)

%matplotlib inline
```

```python
# --------------------------------------------------------------------------- #
# Parameter setting
# No TFP shock
# --------------------------------------------------------------------------- #
A = 1  # Technology level
alpha = 0.36  # Capital share in the Cobb-Douglas production function
beta = 0.95  # Discount factor
psi = 0.25  # Coefficient of leisure in the utility function
theta = 1.5  # Elasticity of leisure in the utility function
```

```python
# --------------------------------------------------------------------------- #
# Analytical solution
# --------------------------------------------------------------------------- #
def ls_compute_analytic(k, alpha=alpha, A=A, psi=psi, theta=theta):
    """ Analytical solution
    Return the optimal labor supply """
    return (((1-alpha) * A * k**alpha) / (psi*theta))**(1 / (theta+alpha-1))


def kplus_compute_analytic(
        k, alpha=alpha, beta=beta, A=A, psi=psi, theta=theta):
    """ Analytical solution
    Return the optimal capital stock in the next period """
    _ls = ls_compute_analytic(k, alpha, A, psi, theta)
    return alpha * beta * A * k**alpha * _ls**(1-alpha)


def c_compute_analytic(k, alpha=alpha, beta=beta, A=A, psi=psi, theta=theta):
    """ Analytical solution
    Return the optimal consumption policy """
    _ls = ls_compute_analytic(k, alpha, A, psi, theta)
    return (1 - alpha*beta)*A*k**alpha*_ls**(1-alpha)


def k_infty_compute(k_infty):
    return k_infty - kplus_compute_analytic(k_infty)


res = root(k_infty_compute, x0=1, method='hybr')
print("Stationary point is {:5f}".format(res.x[0]))
```

```python
# --------------------------------------------------------------------------- #
# Set the exogenous capital domain, must include k_infty
# --------------------------------------------------------------------------- #
kbeg, kend = 0.05, 0.2
gridsize = 250
kgrid = np.linspace(kbeg, kend, gridsize)

# Plot the kplus policy function
plt.plot(kgrid, kplus_compute_analytic(kgrid))
plt.plot(kgrid, kgrid, 'k:')
plt.xlabel(r'$k_{t}$')
plt.ylabel(r'$k_{t_1}$')
plt.xlim([kbeg, kend])
plt.ylim([0, None])
plt.show()

# Plot the labor-supply policy function
plt.plot(kgrid, ls_compute_analytic(kgrid))
plt.xlabel(r'$k_{t}$')
plt.ylabel(r'$n_{t}$')
plt.xlim([kbeg, kend])
plt.ylim([0, None])
plt.show()

# Plot the labor-supply policy function
plt.plot(kgrid, c_compute_analytic(kgrid))
plt.xlabel(r'$k_{t}$')
plt.ylabel(r'$c_{t}$')
plt.xlim([kbeg, kend])
plt.ylim([0, None])
plt.show()
```

## Time iteration collocation with the Gaussian process

Here we aim to numerically solve the neoclassical growth model with the Gaussian process regression. Firstly, we solve the model for the first-order conditions:
\begin{align}
& \lambda_{t} \left(C_{t} - \psi N_{t}^{\theta} \right) - 1 = 0 \\
& \lambda_{t} - \beta \lambda_{t+1} \alpha A K_{t+1}^{\alpha-1}N_{t+1}^{1-\alpha} \\
& A K_{t}^{\alpha}N_{t}^{1-\alpha} - C_{t} - K_{t+1} = 0
\end{align}
where $\lambda_{t}$ is the Lagrangian multiplier associated with the resource constraint.

Hereafter, we solve the system of non-linear equations by using the time iteration collocation. However before doing so, we instantiate and initialize the Gaussian process regressor:

```python
# --------------------------------------------------------------------------- #
# Gaussian process, training and test dataset
# --------------------------------------------------------------------------- #
num_train = 10  # Number of training samples
dim_input = 1  # Dimension of an input of each Gaussian process


def y_train_guess(x_train):
    """ Initialize training outputs """
    _y_train_kplus = x_train
    _y_train_ls = ls_compute_analytic(_y_train_kplus)
    _y_train_c = A * x_train**alpha * _y_train_ls**(1-alpha) - x_train
    _y_train_lambd = 1 / (_y_train_c - psi * _y_train_ls**theta)
    return _y_train_kplus, _y_train_ls, _y_train_c, _y_train_lambd


# Training datasets
x_train_k = np.random.uniform(kbeg, kend, (num_train, dim_input))
y_train_kplus, y_train_ls, y_train_c, y_train_lambd = y_train_guess(x_train_k)

# Check whether each shape is compatible with the GPR class or not
assert x_train_k.shape == (num_train, dim_input), "Shape compatible error"
assert y_train_kplus.shape == (num_train, dim_input), "Shape compatible error"
assert y_train_ls.shape == (num_train, dim_input), "Shape compatible error"
assert y_train_c.shape == (num_train, dim_input), "Shape compatible error"
assert y_train_lambd.shape == (num_train, dim_input), "Shape compatible error"

# Test datasets, used to evaluate the approximation quality
num_test = 1000
x_test_k = np.random.uniform(kbeg, kend, (num_test, dim_input))
```

```python
# --------------------------------------------------------------------------- #
# Instantiate and initialize the Gaussian process
# Use the GPy library and restart hyperparameters optimizations
# --------------------------------------------------------------------------- #
kernel_kplus = GPy.kern.RBF(input_dim=dim_input)
kernel_c = GPy.kern.RBF(input_dim=dim_input)
kernel_lambd = GPy.kern.RBF(input_dim=dim_input)

gp_kplus = GPy.models.GPRegression(x_train_k, y_train_kplus, kernel_kplus)
gp_c = GPy.models.GPRegression(x_train_k, y_train_c, kernel_c)
gp_lambd = GPy.models.GPRegression(x_train_k, y_train_lambd, kernel_lambd)

# Plot without the hyperparameters optimization
gp_kplus.plot()
gp_c.plot()
gp_lambd.plot()
```

```python
# Fit to data using Maximum Likelihood Estimation of the parameters
n_restarts = 10

gp_kplus.optimize_restarts(n_restarts, verbose=False)
gp_c.optimize_restarts(n_restarts, verbose=False)
gp_lambd.optimize_restarts(n_restarts, verbose=False)

# Plot with the hyperparameters optimization
gp_kplus.plot()
gp_c.plot()
gp_lambd.plot()
```

```python
# --------------------------------------------------------------------------- #
# Equilibrium conditions
# --------------------------------------------------------------------------- #
def euler(x0, k, gp_lambd):
    """
    Set of non-linear equilibrium conditions solved by IPOPT
    x0: Starting values for the optimization
    k: Current capital state
    x[0]: Capital stock in the next period
    x[1]: Consumption
    x[2]: lambda, Lagrange multiplier associated with the budget constraint
    """

    nvar = 3  # Number of variables

    # All of plicies are assumed to be non-negative
    x_L = np.zeros(nvar, dtype=float)
    x_U = np.ones(nvar, dtype=float) * np.inf

    def lambdplus(x):
        """ Return lambda (the Lagrange multiplier) in the next period """
        assert len(x) == nvar
        return gp_lambd.predict(x[0].reshape(-1, dim_input))[0].ravel()

    def ls_plus(x):
        """ Labor supply in the next period """
        assert len(x) == nvar
        return ls_compute_analytic(x[0])

    def eval_f(x):
        """ Dummy objective function """
        assert len(x) == nvar
        return 0

    def eval_grad_f(x):
        "Gradient of the dummy objective function"
        assert len(x) == nvar
        grad_f = np.zeros(nvar, dtype=float)
        return grad_f

    ncon = nvar  # Number of constraints

    """ Complementarity constraints
    g0: FOC wrt. consumption
    g1: Euler equation wrt. k_{t+1}
    g2: Resource constraint """

    g_L = np.zeros(ncon, dtype=float)
    g_U = g_L

    nnzj = int(nvar * ncon)  # Number of (possibly) non-zeros in Jacobian
    nnzh = int((nvar**2 - nvar) / 2 + nvar)  # Number of non-zeros in Hessian

    ls = ls_compute_analytic(k)  # Current labor supply

    def eval_g(x):
        """ The system of non-linear equilibrium conditions """
        assert len(x) == nvar

        g0 = 1 - x[2] * (x[1] - psi * ls**theta)
        g1 = beta * lambdplus(x) * A * alpha * x[0]**(alpha-1) * ls_plus(x)**(
            1-alpha) - x[2]
        g2 = A * k**alpha * ls**(1-alpha) - x[0] - x[1]
        return np.array([g0, g1, g2])

    def eval_jac_g(x, flag):
        """ Numerical approximation of the Jacobian of the system of
        non-linear equilibrium conditions
        Use the finite-difference-values option provided by IPOPT """
        assert len(x) == nvar

        row_idx = np.empty(nnzj, dtype=int)  # Row index
        col_idx = np.empty(nnzj, dtype=int)  # Column index

        # Jacobian matrix structure
        if flag:
            for i in range(ncon):
                for j in range(nvar):
                    row_idx[j + i * nvar] = i
                    col_idx[j + i * nvar] = j

            return (row_idx, col_idx)

    # ----------------------------------------------------------------------- #
    # Define a NLP model
    # ----------------------------------------------------------------------- #
    pyipopt.set_loglevel(0)  # Let IPOPT quite

    neoclassical = pyipopt.create(
        nvar, x_L, x_U, ncon, g_L, g_U, nnzj, nnzh, eval_f, eval_grad_f,
        eval_g, eval_jac_g)
    neoclassical.str_option("linear_solver", "ma57")
    neoclassical.str_option(
        "jacobian_approximation", "finite-difference-values")
    neoclassical.str_option("hessian_approximation", "limited-memory")
    neoclassical.int_option("print_level", 1)

    xstar, zl, zu, constraint_multipliers, obj, status = neoclassical.solve(x0)

    if status != 0:
        warnings.warn(
            "IPOPT fails to solve the system of non-linear equations. "
            "Use the starting value as the solution instead.")
        xstar = x0
    return xstar
```

```python
# --------------------------------------------------------------------------- #
# Time iteration collocation
# --------------------------------------------------------------------------- #
def time_iter_gpr(num_train, dim_input):
    """
    Time iteration collocation with the Gaussian process regression
    num_train: Number of training examples
    dim_input: Number of an input dimension
    """
    num_iter = 50  # Number of time iterations
    epsilons = []  # Track the approximation error
    epsilon_tol = 1e-5  # Convergence tolrance

    # ----------------------------------------------------------------------- #
    # Generate a training dataset
    # ----------------------------------------------------------------------- #
    x_train_k = np.random.uniform(kbeg, kend, (num_train, dim_input))
    y_train_kplus, y_train_ls, y_train_c, y_train_lambd = y_train_guess(
        x_train_k)

    # ----------------------------------------------------------------------- #
    # Instantiate and initialize the Gaussian process
    # ----------------------------------------------------------------------- #
    kernel_kplus = GPy.kern.RBF(input_dim=dim_input)
    kernel_c = GPy.kern.RBF(input_dim=dim_input)
    kernel_lambd = GPy.kern.RBF(input_dim=dim_input)

    gp_kplus = GPy.models.GPRegression(x_train_k, y_train_kplus, kernel_kplus)
    gp_c = GPy.models.GPRegression(x_train_k, y_train_c, kernel_c)
    gp_lambd = GPy.models.GPRegression(x_train_k, y_train_lambd, kernel_lambd)

    # Fit to data using Maximum Likelihood Estimation of the parameters
    n_restarts = 10

    gp_kplus.optimize_restarts(n_restarts, verbose=False)
    gp_c.optimize_restarts(n_restarts, verbose=False)
    gp_lambd.optimize_restarts(n_restarts, verbose=False)

    # ----------------------------------------------------------------------- #
    # Time iteration collocation
    # ----------------------------------------------------------------------- #
    for n in range(1, num_iter+1):

        # Starting value retliving from the previous optimization
        x0 = np.stack(
            [y_train_kplus.ravel(), y_train_c.ravel(), y_train_lambd.ravel()],
            axis=1)

        # Keep the optimal policies
        y_train_kplus = np.empty((num_train, dim_input), dtype=float)
        y_train_c = np.empty_like(y_train_kplus)
        y_train_lambd = np.empty_like(y_train_kplus)

        for idx, k in enumerate(x_train_k):
            # For each state, solve the system of non-linear equations
            xstar = euler(x0[idx, :], k, gp_lambd)
            # Track the optimal policies
            y_train_kplus[idx, 0] = xstar[0]
            y_train_c[idx, 0] = xstar[1]
            y_train_lambd[idx, 0] = xstar[2]

        # sys.exit(0)
        # ------------------------------------------------------------------- #
        # Train the Gaussian process with the optimal policy
        # ------------------------------------------------------------------- #
        # Capital stock in the next period
        kernel_kplus_updated = GPy.kern.RBF(dim_input)
        gp_kplus_updated = GPy.models.GPRegression(
            x_train_k, y_train_kplus, kernel_kplus_updated)
        gp_kplus_updated.optimize_restarts(n_restarts, verbose=False)

        # Consumption policy
        kernel_c_updated = GPy.kern.RBF(dim_input)
        gp_c_updated = GPy.models.GPRegression(
            x_train_k, y_train_c, kernel_c_updated)
        gp_c_updated.optimize_restarts(n_restarts, verbose=False)

        # Lagrange multiplier
        kernel_lambd_updated = GPy.kern.RBF(dim_input)
        gp_lambd_updated = GPy.models.GPRegression(
            x_train_k, y_train_lambd, kernel_lambd_updated)
        gp_lambd_updated.optimize_restarts(n_restarts, verbose=False)

        # ------------------------------------------------------------------- #
        # Approximation error analysis
        # Update the policy functions for the next iteration
        # ------------------------------------------------------------------- #
        epsilon = max(
            np.max(np.abs(gp_kplus.predict(x_test_k)[0]
                          - gp_kplus_updated.predict(x_test_k)[0])),
            np.max(np.abs(gp_c.predict(x_test_k)[0]
                          - gp_c_updated.predict(x_test_k)[0])),
            np.max(np.abs(gp_lambd.predict(x_test_k)[0]
                          - gp_lambd_updated.predict(x_test_k)[0]))
            )
        epsilons.append(epsilon)  # Track the history of epsilon

        if n % 1 == 0:
            print('Iteration: {}, Approximation error: {:.3e}'.format(
                n, epsilons[-1]))

        if epsilons[-1] < epsilon_tol:
            # Terminate the time iteration and save the optimal surrogates
            gp_kplus_star = gp_kplus_updated
            gp_c_star = gp_c_updated
            gp_lambd_star = gp_lambd_updated
            print("Time iteration collocation is terminated successfuly with "
                  "{} iterations".format(n))
            break  # Terminate the iteration

        else:
            # Update the GP with the surrogates
            gp_kplus = gp_kplus_updated
            gp_c = gp_c_updated
            gp_lambd = gp_lambd_updated

    return epsilons, gp_kplus_star, gp_c_star, gp_lambd_star
```

Now suppose that there are 10 training samples:

```python
epsilons, gp_kplus_star, gp_c_star, gp_lambda_star = time_iter_gpr(
    num_train=10, dim_input=1)
```

```python
# --------------------------------------------------------------------------- #
# Plot
# --------------------------------------------------------------------------- #
# Approximation error
plt.plot(epsilons)
plt.yscale('log')
plt.xlabel("Number of iterations")
plt.ylabel("Max approximation error")
plt.xlim([0, None])
plt.show()

# Capital stock in the next period
kgrid = kgrid.reshape(gridsize, dim_input)
kplus_star = gp_kplus_star.predict(kgrid)
mu_kstar = kplus_star[0]
sigma_kstar = np.sqrt(kplus_star[1])
plt.plot(kgrid, mu_kstar, 'k-', label=r'$\mu$')
plt.plot(kgrid, mu_kstar + 2*sigma_kstar, 'r-.', label=r'$\mu + 2\sigma$')
plt.plot(kgrid, mu_kstar - 2*sigma_kstar, 'r:', label=r'$\mu + 2\sigma$')
plt.plot(kgrid, kplus_compute_analytic(kgrid), 'b--', label='True')
plt.plot(kgrid, kgrid, 'k:')
plt.xlim([kbeg, kend])
plt.xlabel(r"$k_{t}$")
plt.ylim([0, None])
plt.ylabel(r"$k_{t+1}$")
plt.legend(loc='upper left')
plt.show()

# Consumption
c_star = gp_c_star.predict(kgrid)
mu_cstar = c_star[0]
sigma_cstar = np.sqrt(c_star[1])
plt.plot(kgrid, mu_cstar, 'k-', label=r'$\mu$')
plt.plot(kgrid, mu_cstar + 2*sigma_cstar, 'r-.', label=r'$\mu + 2\sigma$')
plt.plot(kgrid, mu_cstar - 2*sigma_cstar, 'r:', label=r'$\mu + 2\sigma$')
plt.plot(kgrid, c_compute_analytic(kgrid), 'b--', label='True')
plt.xlim([kbeg, kend])
plt.xlabel(r"$k_{t}$")
plt.ylim([0, None])
plt.ylabel(r"$c_{t}$")
plt.legend(loc='upper left')
plt.show()
```

It must be interesting to perform a sensitivity analysis with respect to the number of training examples.

```python
num_trains = [4, 6, 8, 10]
epsilons_list = [''] * len(num_trains)
gp_kplus_star_list = [''] * len(num_trains)
gp_c_star_list = [''] * len(num_trains)
gp_lambd_star_list = [''] * len(num_trains)

for idx, num_train in enumerate(num_trains):
    print("Number of training examples are {}".format(num_train))
    epsilons_list[idx], gp_kplus_star_list[idx], gp_c_star_list[idx], \
    gp_lambd_star_list[idx] = time_iter_gpr(num_train, 1)
```
